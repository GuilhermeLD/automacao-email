```python
import yfinance
```

```python
codigo = input("Digite o código da ação desejada: ")

dados = yfinance.Ticker(codigo).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

## Gerando as análises de forma automática

- Análise dos últimos 6 meses
- Cotação máxima
- Cotação Mínima
- Cotação Atual

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[0]
```

```python
import pyautogui
import pyperclip
```

```python
# Setando o intevalo dos comandos
pyautogui.PAUSE = 3

# Abrir uma nova aba no naveggador (ctrl + t)
pyautogui.hotkey("ctrl", "t")

# digitar o endereco do yahoo e dar enter
pyperclip.copy("https://mail.yahoo.com/")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# clicar no botao escrever
pyautogui.click(x=114, y=214)

# colocar email // precisei colocar essa linha de esc e tab a mais, por conta da sugestão do yahoo
pyperclip.copy("guilherme_limadavid@yahoo.com.br")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")
pyautogui.hotkey("esc")
pyautogui.hotkey("tab")

# colocar assunto
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

mensagem = f"""
Prezador Gestor,

Seguem as análises diárias dos últimos seis meses da ação {codigo}:

Cotação máximo: R$ {round(maxima,2)}
Cotação Mínima: R$ {round(minima,2)}
Cotação Atual: R$ {round(atual,2)}

Qualquer duvida, sigo a disposição!

"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# clicar no botão de enviar
pyautogui.click(x=1028, y=990)
```
